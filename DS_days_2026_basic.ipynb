```python
import requests
import pandas as pd

# Query URL (adjust parameters as needed)
url = "https://earthquake.usgs.gov/fdsnws/event/1/query"
params = {
    "format": "geojson",
    "orderby": "time",
    "limit": 1500  # number of events
}

# Fetch the GeoJSON
resp = requests.get(url, params=params)
data = resp.json()

```

```python
data["metadata"]
```

```python
event = data["features"][0]
print(event.keys())
```

```python
print(event["geometry"])
```

```python
print(event["properties"].keys())
```

```python
# Extract coordinates and magnitudes
features = data["features"]
lons = [feat["geometry"]["coordinates"][0] for feat in features]
lats = [feat["geometry"]["coordinates"][1] for feat in features]
mags = [feat["properties"]["mag"] for feat in features]

# Put into a DataFrame
df = pd.DataFrame({
    "longitude": lons,
    "latitude": lats,
    "magnitude": mags
})

df.to_csv(
    "earthquakes.txt",
    sep=" ",
    index=False,
    header=False
)
```

```python
import pygmt

region = 'd'
proj = 'N15c'
fig = pygmt.Figure()
pygmt.makecpt(cmap = 'geo')
fig.grdimage(grid=grid, 
            region = region,
            projection = proj,
            cmap='geo')
# Set up global map
fig.coast(
    region=region, 
    projection=proj,  
    shorelines=True,
    frame=True
)
fig.colorbar(
    frame='af+l"Elevation (m)"'
)

# Plot earthquakes
# fig.plot(
#     x=df["longitude"], y=df["latitude"],
#     #style="c0.2c",
#     #color="red",
#     style='c',
#     fill = 'red',
#     size = 0.1 * df['magnitude'],
#     pen="black"
# )

fig.plot(
    data="earthquakes.txt",
    style="c0.2c",
    #size="0.0001",     
    fill="red",
    pen="black",
    #size="0.1", 
)


fig.show()
```
