```python
import pygmt
fig = pygmt.Figure()
fig.coast(shorelines=True,
         region = 'd',
         projection = 'Q0/0/12c',
         land = 'blue',
          water='green',
         frame=["xa30g30", "ya30g30", "WSen"]

         )

fig.plot(
    x=[-2, 10],
    y=[66, -45],
    style="c0.3c",     # circle, 0.3 cm
    #style = 'c',
    fill="red",
    pen="1p,black",
)

# Add text label next to the circle
fig.text(
    x=10.75,
    y=59.91,
    text="Oslo",
    offset="0.3c/0.2c",   # move label right & up
    font="10p,Helvetica-Bold,white",
    justify="LM",         # Left-Middle alignment
)
fig.show()
```

```python
import pygmt

fig = pygmt.Figure()
fig.coast(
    region="d",
    #projection="Q0/0/12c",
    projection="G10/60/12c",   # lon0/lat0/size
    land="blue",
    water="green",
    shorelines=True,
    borders="1/0.8p,white",   #  country borders in white
    dcw="NO+gyellow", 
    frame=["xa30g30", "ya30g30", "WSen"],
)
fig.show()
```
